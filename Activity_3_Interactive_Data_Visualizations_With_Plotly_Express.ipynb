```python
# Loading the modules
import pandas as pd
import seaborn as sns
import plotly.express as px
from numpy import median, mean
```

```python
#https://github.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/tree/master/datasets

#reading the data sets .csv
path1 = '/content/drive/MyDrive/Activities/co2.csv'
path2='/content/drive/MyDrive/Activities/gapminder.csv'
co2= pd.read_csv(path1)
gm= pd.read_csv(path2)
```

```python
co2.head()
```

```python
gm.head()
```

```python
#Transformation
#Levels of c02 in each region

location = gm[['Country', 'region']].drop_duplicates()
location_w_regions = pd.merge(co2, location, left_on='country', right_on='Country', how='inner')
location_w_regions = location_w_regions.drop('Country', axis='columns')
new_co2 = pd.melt(location_w_regions, id_vars=['country', 'region'])
columns = ['country', 'region', 'year', 'co2']
new_co2.columns = columns
df_co2 = new_co2[new_co2['year'].astype('int64') > 1963]
df_co2 = df_co2.sort_values(by=['country', 'year'])
df_co2['year'] = df_co2['year'].astype('int64')
df_gdp = gm[['Country', 'Year', 'gdp']]
df_gdp.columns = ['country', 'year', 'gdp']
data = pd.merge(df_co2, df_gdp, on=['country', 'year'], how='left')
data = data.dropna()
```

```python
# scatter plot (x,y,year,co2) - let us to  determine the color of the regions of the datapoints
scat = px.scatter(data, x = 'year', y = 'co2', color = 'region', marginal_y = 'box')
```

```python
scat.show()
```

```python
#Added the gdp
scat1 = px.scatter(data, x = 'gdp', y = 'co2', color = 'region', marginal_y = 'box', marginal_x = 'rug', animation_frame = 'year', animation_group = 'country')
scat1.show()
```

```python
#Displayed the density of the areas through years
dens1 = px.density_contour(data, x="gdp", y="co2", color="region", marginal_x="rug", marginal_y="histogram", animation_frame = 'year', animation_group = 'region')
```
