```python
# Loading the modules
import pandas as pd
import seaborn as sns
from numpy import median, mean
```

```python
#https://github.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/datasets

#reading the data set Athletes.csv
path = '/content/drive/MyDrive/Activities/athlete_events.csv'
olym_athletes= pd.read_csv(path)
olym_athletes.head()
```

```python
# Medal winners
olym_athletes_winners = olym_athletes.dropna(subset=['Medal'])
olym_athletes_winners.head()
```

```python
#winners of 2016 Promotion
winners_2016 = olym_athletes_winners[(olym_athletes_winners.Year ==
2016)]
winners_2016.head()
```

```python
#Top Sports
top_sports=['Athletics','Swimming','Rowing','Football', 'Hockey']
winners_2016.Sport.value_counts()


```

```python

#Winners in 2016 in top Sports
top_sports_winners_2016 = winners_2016[(winners_2016.Sport.isin(top_sports))]
top_sports_winners_2016.head()
```

```python
top_sports_winners_2016[['Age', 'Height', 'Weight']]
```

```python
#global pattern and correlation betweeen the H and W of athletes

# import the seaborn library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
figure1 = plt.figure()
ax = figure1.add_subplot(111)
ax = sns.scatterplot(x="Height", y="Weight", data=top_sports_winners_2016)
plt.show()
```

```python
sns.jointplot(top_sports_winners_2016.Height,  kind="hex", color="gray")
```

```python
sns.set_theme(style="ticks")
sns.jointplot(top_sports_winners_2016.Weight,   kind="hex")
```

```python
#joinplot correlation
## hexbin plot

sns.set_theme(style="ticks")
sns.jointplot(x=top_sports_winners_2016.Height, y=top_sports_winners_2016.Weight, kind="hex", color="#4CB391")
```

```python
#plot coralation within medal and weigth
sns.set_style('white')
axis1 = sns.violinplot(x='Medal', y='Weight', data=top_sports_winners_2016, hue='Sex')
```

```python
#plot coralation within medal and weigth
axis2 = sns.violinplot(x='Medal', y='Height', data=top_sports_winners_2016, hue='Sex')
```
