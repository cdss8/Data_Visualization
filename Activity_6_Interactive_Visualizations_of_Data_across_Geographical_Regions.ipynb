```python
# Loading the modules
import pandas as pd
import seaborn as sns
import plotly.express as px
from numpy import median, mean
```

```python
#https://github.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/datasets

#reading the data set .csv
path ='//content/drive/MyDrive/Activities/share-of-electricity-production-from-renewable-sources.csv'
ren_energy= pd.read_csv(path)
ren_energy.head()
```

```python
ren_energy.sort_values(by=['Year'],inplace=True)
ren_energy.head()
```

```python
#Generating a choropleth map using the plotly express module animated based on Year
import plotly.express as px

ren_energy = ren_energy.query('Year<2017 and Year>2007')

fig = px.choropleth(ren_energy, locations="Code", color="Renewable electricity (% electricity production)",
hover_name="Country", animation_frame="Year", color_continuous_scale='Greens')
```

```python
# drawing
fig.update_layout(title_text='Renewable energy production across the world (% of electricity production)',
                 geo=dict(projection={'type': 'natural earth'}, projection_type='orthographic'))

fig.show()
```
